```python
import pandas as pd
import numpy as np
import xgboost as xgb
import redis
import requests
import os
from dotenv import load_dotenv
from kafka import KafkaProducer

load_dotenv('../.env')

# Test OpenWeatherMap API
api_key = os.getenv('OPENWEATHER_API_KEY')
url = f"http://api.openweathermap.org/data/2.5/weather?q=Delhi&appid={api_key}&units=metric"
response = requests.get(url)
data = response.json()

print("OpenWeatherMap API working!")
print(f"Delhi temperature:  {data['main']['temp']}°C")
print(f"Condition:          {data['weather'][0]['description']}")
print(f"Humidity:           {data['main']['humidity']}%")

# Test Redis
r = redis.Redis(host='localhost', port=6379, db=0)
r.set('test', 'working')
print(f"\nRedis working: {r.get('test').decode()}")

# Test Kafka
producer = KafkaProducer(
    bootstrap_servers='localhost:9092',
    api_version=(2, 8, 0)
)
producer.close()
print("Kafka working!")

print(f"\nXGBoost version: {xgb.__version__}")
print("\nAll systems ready — Step 1 complete!")
print("Ready for Step 2!")
```
